```python
import random
import time

POPULATION_SIZE = 100
SUBJECTS_SIZE = 6
PROFESSORS_SIZE = 3
PERIODS = 6
DAYS = 5

SUBJECTS = "ABCDEF"
PROFESSORS = "xyz"
MAP = "yxxzyx"

SUB = [7, 3, 6, 4, 5, 5]
PROF = [13, 12, 5]


def random_num(start, end):
    return random.randint(start, end)


def mutated_gene():
    return random.choice(SUBJECTS)


def create_gnome():
    len_gnome = PERIODS * DAYS
    gnome = ""
    for _ in range(len_gnome):
        gnome += mutated_gene()
    return gnome


class Individual:
    def __init__(self, chromo):
        self.chromosome = chromo
        self.fitness = self.calc_fitness()

    def mate(self, parent2):
        child_chromosome = ""
        for i in range(len(self.chromosome)):
            p = random_num(0, 100) / 100
            if p < 0.45:
                child_chromosome += self.chromosome[i]
            elif p < 0.90:
                child_chromosome += parent2.chromosome[i]
            else:
                child_chromosome += mutated_gene()
        return Individual(child_chromosome)

    def conflict_sub_hours(self):
        count = 0
        conflicts = 0
        for i in range(SUBJECTS_SIZE):
            count = 0
            sub = SUBJECTS[i]
            for j in range(len(self.chromosome)):
                if self.chromosome[j] == sub:
                    count += 1
            conflicts += abs(count - SUB[i])
        return conflicts

    def conflict_sub_cont(self):
        conflicts = 0
        for i in range(len(self.chromosome) - 1):
            if self.chromosome[i] == self.chromosome[i + 1]:
                conflicts += 1
        return conflicts

    def conflict_prof_hours(self):
        conflicts = 0
        for i in range(PROFESSORS_SIZE):
            count = 0
            prof = PROFESSORS[i]
            for j in range(len(self.chromosome)):
                if prof == MAP[SUBJECTS.index(self.chromosome[j])]:
                    count += 1
            conflicts += abs(count - PROF[i])
        return conflicts

    def calc_fitness(self):
        fitness = 0
        fitness += self.conflict_sub_hours()
        fitness += self.conflict_sub_cont()
        fitness += self.conflict_prof_hours()
        return fitness


def display(population):
    print("display")
    for i in range(len(population[0].chromosome)):
        if i % PERIODS == 0:
            print()
        print("\t", population[0].chromosome[i], end="")
    print("\n\n\tfitness :", population[0].fitness)


def main():
    random.seed(time.time())
    population = []
    found = False
    for _ in range(POPULATION_SIZE):
        gnome = create_gnome()
        population.append(Individual(gnome))
    generation = 0
    count = 0

    while not found:
        population.sort(key=lambda x: x.fitness)

        if population[0].fitness <= 0 or count == 500:
            found = True
            break

        new_generation = population[:10]

        for _ in range(int(0.9 * POPULATION_SIZE)):
            r = random_num(0, 49)
            parent1 = population[r]
            r = random_num(0, 49)
            parent2 = population[r]
            offspring = parent1.mate(parent2)
            new_generation.append(offspring)

        new_generation.sort(key=lambda x: x.fitness)

        if new_generation[0].fitness == population[0].fitness:
            count += 1
        else:
            count = 0

        population = new_generation
        generation += 1

    display(population)


if __name__ == "__main__":
    main()
```
